```python
#Code by Deven Bibikar
#Written in Python
#Updated as of July 2021
#Program intended to grab a large list of all chemical compounds and allow sorting compounds by chemicals.

import pandas as pd

url = 'https://en.wikipedia.org/wiki/Glossary_of_chemical_formulae'

all_tables = pd.read_html(url)

#print(all_tables)

all_tables.pop()
#print(len(all_tables))

element_frame = []

for list_item in all_tables:
    for formula in list_item["Chemical formula"]:
        element_frame.append(formula)
    
#print(element_frame)

askElement = []
continue_search = False

test_elements = []

#my_input = input('Enter an element to search for:')


while True:
    my_input = input('Enter an element to search for:')
    #print(my_input)
    #condition1 = ("q" or "quit") in my_input.lower()
    #print(condition1)
    
    if ("q" or "quit") in my_input.lower():
        break
        
    elif ("d" or "done") in my_input.lower():
        continue_search = True
        break
        
    else:
        askElement.append(my_input)

print(askElement)



if continue_search:
    
    for item in element_frame:
        if askElement[0] in item:
            test_elements.append(item)
    
     
    for item in test_elements:
        for query in askElement:
            if query not in item:       
                try:
                    test_elements.remove(item)
                    print("Removing")
                    continue
                    
                except ValueError:     
                    continue
                
                
    print(test_elements)
        
        
        

    
    
```
